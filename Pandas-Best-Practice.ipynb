# 8. Which occupations deliver the funniest TED talks on average?

Bonus exercises:

- for each talk, calculate the most frequent rating
- for each talk, clean the occupation data so that there's only one occupation per talk

```python
ted.ratings_list[0]
```

```python
# Count the number of Funny rating

ted.ratings.str.contains('Funny').value_counts()
```

```python
ted.
```

```python
def get_funny_ratings(list_of_dict):
    
    for d in list_of_dict:
        if d['name'] == 'Funny':
            return d['count']
```

```python
get_funny_ratings(ted.ratings_list[0])
```

```python
ted['funny_ratings'] = ted.ratings_list.apply(get_funny_ratings)
ted.head()
```

```python
ted['funny_rate'] = ted.funny_ratings / ted.num_ratigns
```

```python
#Checking if the funny_rate make sence comparing to the speaker occupation
ted.sort_values('funny_rate', ascending=False).speaker_occupation.head(20)
```

```python
ted.groupby('speaker_occupation').funny_rate.mean().sort_values(ascending=False).head()
```

```python
ted.speaker_occupation.describe()
```

### Step 5: Re-analyze the funny rate by occupation (for top occupations only)

```python
occupation_counts = ted.speaker_occupation.value_counts()
```

```python
type(occupation_counts)
```

```python
occupation_counts[occupation_counts > 5]
```

```python
#Extracting the index of the occupation that have 
top_occupation = occupation_counts[occupation_counts > 5].index
top_occupation
```

```python
ted_top_occupation = ted[ted.speaker_occupation.isin(top_occupation)]
ted_top_occupation.shape
```

```python
ted_top_occupation.head(3)
```

```python
ted_top_occupation.groupby('speaker_occupation').funny_rate.agg(['count','mean']).sort_values('mean',ascending=False)
```

```python
ted_top_occupation.groupby('speaker_occupation').funny_rate.mean().sort_values(ascending=False).plot(kind='bar', figsize=(20,15))
```
